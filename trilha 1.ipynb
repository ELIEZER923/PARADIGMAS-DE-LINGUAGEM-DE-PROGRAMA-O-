```python
import urllib.request  

cidade = []

for line in urllib.request.urlopen('https://drive.google.com/file/d/1nQmgNvOQZygLxvsYnHYCaf0M0bveOMSS/view?usp=sharing'): 
    cidade.append(line.decode('utf-8'))

for i in range(len(cidade)):
    cidade[i] = cidade[i].lower()
    cidade[i] = cidade[i].replace('\n','')
    cidade[i] = cidade[i].replace('.','')
    cidade[i] = cidade[i].replace(',','')
    cidade[i] = cidade[i].replace('(','')
    cidade[i] = cidade[i].replace(')','')
    cidade[i] = cidade[i].replace('?','')
    cidade[i] = cidade[i].replace('\'','') 
```

```python
mydict = {}                            

for line in cidade:

  line = line.lower()                  
  words = line.split()                

  for word in words:                     
    if word not in mydict.keys():         
      mydict[word] = 0.5                   
    else:                                 
      mydict[word] =  mydict[word] + 0.5
print(mydict)    
```

```python
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt

df = pd.DataFrame(mydict.items(),  columns=['palavras', 'count']).sort_values('count',ascending=False)
df = df[df['count'] > 8]


plt.figure(figsize=(23,15))
sns.barplot(x=df.palavras,y=df['count'])
plt.xticks(rotation=70) 

plt.show()
```
